```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def load_application_train():
    data = pd.read_csv("application_train.csv")
    return data 

def load_titanic():
    data = pd.read_csv("titanic.csv")
    return data 

df = load_titanic()
df.head()
```

```python
df["new_cabin_flag"] = np.where(df["Cabin"].isnull(),0,1)
df["new_cabin_flag_2"] = df["Cabin"].notnull().astype(int)
df.head()
```

```python
df.groupby("new_cabin_flag").agg({"Survived":["mean","count"]})
```

```python
from statsmodels.stats.proportion import proportions_ztest
stat,pvalue = proportions_ztest(count=[df.loc[df["new_cabin_flag"]==1,"Survived"].sum(),
                                       df.loc[df["new_cabin_flag"]==0,"Survived"].sum()],
                                nobs=[df.loc[df["new_cabin_flag"]==1,"Survived"].shape[0],
                                      df.loc[df["new_cabin_flag"]==0,"Survived"].shape[0]])
print(f"pval = {pvalue:.10f}")
```

```python
df.loc[(df["SibSp"]+df["Parch"])>0,"is_alone"] = 0
df.loc[(df["SibSp"]+df["Parch"])==0,"is_alone"] =1
df.head()
```

```python
df.groupby("is_alone").agg({"Survived":["mean","count"]})
```

```python
stat,pvalue = proportions_ztest(count=[df.loc[df["is_alone"]==1,"Survived"].sum(),
                                       df.loc[df["is_alone"]==0,"Survived"].sum()],
                                nobs=[df.loc[df["is_alone"]==1,"Survived"].shape[0],
                                      df.loc[df["is_alone"]==0,"Survived"].shape[0]])
print(f"pval = {pvalue:.10f}")
```
